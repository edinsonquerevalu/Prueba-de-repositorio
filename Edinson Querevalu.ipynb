```python
"""
Haga un programa que sume todos los números dados por un usuario (use for)
"""
print('Suma de números enteros dados por un usuario')
lista_numeros = []
decision ='S'
while decision == 'S':
  numero=int(input('Ingrese el siguiente número: '))
  lista_numeros.append(numero)
  decision = input('Escribir "S" (mayúscula) para añadir otro número: ')
else: 
    print('Su lista de números es ',lista_numeros)
suma=0
for i in lista_numeros:
  suma=suma+i
print('La sumatoria de todos los números de la lista es ',suma)
```

```python
"""
Hacer una función para guardar emails de clientes en un excel 
"""

import pandas as pd
def guardaremail(): 
  nombresclientes=[]
  emailsclientes=[]
  decision='S'
  while decision == 'S':
    _nombrecliente=input('Ingrese nombre del cliente: ')
    _emailcliente=input('Ingrese email del cliente: ')
    nombresclientes.append(_nombrecliente)
    emailsclientes.append(_emailcliente)
    decision=input('Ingrese S para añadir otro cliente: ')
  else:
      print('No se desea añadir datos de más clientes, archivo guardado como "email clientes.xlsx"')
      dicc_clientes_email={"Nombre":nombresclientes,"Email":emailsclientes}
      usuarios_df=pd.DataFrame(dicc_clientes_email)
      usuarios_df.to_excel("email clientes.xlsx")
      
guardaremail()
```

```python
"""
Hacer un programa para almacenar datos de clientes (celular, telefono, email) en un excel
"""

import pandas as pd

def almacenadatos():
  nombresclientes=[]
  emailsclientes=[]
  telefonosclientes=[]
  celularesclientes=[]
  decision='S'
  while decision=='S':
    _nombres=input('Ingresar nombre del cliente: ')
    _celular=input('Ingresar celular del cliente: ')
    _telefono=input('Ingresar teléfono del cliente: ')
    _email=input('Ingresar email del cliente: ')
    nombresclientes.append(_nombres)
    emailsclientes.append(_email)
    celularesclientes.append(_celular)
    telefonosclientes.append(_telefono)
    decision=input('Escriba S para agregar los datos de otro cliente: ')
  else: 
      print('No desea agregar datos de más clientes, archivo guardado como "datosclientes.xlsx"')
      dicc_clientes={"Nombre":nombresclientes, "Celular":celularesclientes, "Teléfono":telefonosclientes, "email":emailsclientes}
      datosclientes_df=pd.DataFrame(dicc_clientes)
      datosclientes_df.to_excel("datosclientes.xlsx")
almacenadatos()
```

```python
# Genere un función para calcular la transpuesta de una matriz

import numpy as np
matriz = []
def transpuesta():
    nc = int(input('Ingresar número de columnas: '))
    nf = int(input('Ingresar número de filas: '))
    for i in range(nf):
      matriz.append([0]*nc)
    for j in range(nf):
      for k in range(nc):
        matriz[j][k] = int(input('Posición %d,%d: '%(j,k)))

  m = np.array(matriz)
  transpuesta = m.transpose()
  
  print('Matriz transpuesta: ')
  print(transpuesta)
transpuesta()
```

```python
# Genere una función para encontrar el producto interno de un vector por si mismo
import numpy as np
mat=[]
def prodpunto():
    filas=1
    longitud=int(input('Ingresar longifud del vector: '))
    for i in range(filas):
      mat.append([0]*longitud)
    for j in range(filas):
      for b in range(longitud):
        mat[j][b] = int(input('Elemento %d,%d: '%(j,b)))
  
    x=np.array(mat)
    y=x[0]
    prod=y.dot(y)
    print('El producto punto del vector por sí mismo es: ',prod)
prodpunto()


```
